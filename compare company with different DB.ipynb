```python
df2 = df.copy()
```

```python
df2['Target Company'] = df2['Target Company'].str.strip()
```

```python
df2['Target Company'] = df2['Target Company'].str.rsplit("(")
```

```python
for i in range(len(df2['Target Company'])):
    if len(df2['Target Company'][i]) > 1:
        df2['Target Company'][i] = df2['Target Company'][i][0]
```

```python
def flatten(potential_list): 
    new_list = [] 
    for e in potential_list: 
        if isinstance(e, list): 
            new_list.extend(flatten(e)) 
        else: 
            new_list.append(e) 
    return new_list 
```

```python
target_company = flatten(target_company)           
```
